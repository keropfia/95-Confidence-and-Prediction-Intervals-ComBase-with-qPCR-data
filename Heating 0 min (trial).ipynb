```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# line 1 points
x1 = [0,1,2,3,4]
y1 = [10,10,10,10,10]

# line 2 points
x2 = [0,1,2,3,4]
y2 = [10,9,9,10,10]

# plotting the points
plt.plot(x1, y1, color='red', linestyle='dashed', linewidth = 1, marker='o', markerfacecolor='yellow', markersize=6, label='BPW')
plt.plot(x2, y2, color='green', linestyle='dashed', linewidth = 1, marker='o', markerfacecolor='blue', markersize=6, label='D-BPW')

# setting x and y axis range
plt.ylim(0,12)
plt.xlim(0,4)

# naming the x axis
plt.xlabel('Time (h)')
# naming the y axis
plt.ylabel('No. of positive samples')

# giving a title to my graph
plt.title('0 min heating')

# filling in between line
plt.fill_between(x2, np.maximum(y1, y2), color='grey', alpha=0.5)
plt.fill_between(x1, np.minimum(y1, y2), [0] * len(x1), color='cyan', alpha=0.5)

# function to show the plot
plt.show()
```

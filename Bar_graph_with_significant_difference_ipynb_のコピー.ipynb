```python
import numpy as np
import matplotlib.pyplot as plt

# Data
means = [11.8462097013076, 7.87070428186792]
stds = [0.598274786868603, 0.43041556538705]

labels = ['0 h', '6 h']
x = np.array([0.0, 0.8])

# Create figure
fig, ax = plt.subplots(figsize=(5, 5))

# Bar plot
ax.bar(
    x,
    means,
    yerr=stds,
    capsize=5,
    width=0.4,
    color=['white'],edgecolor='black',
    linewidth=1.5,
    error_kw={
        'ecolor': 'black',
        'elinewidth': 1.5
    }
)

# Function to add significance bracket
def add_significance(ax, x1, x2, y, h, text):
    ax.plot(
        [x1, x1, x2, x2],
        [y, y+h, y+h, y],
        lw=1.5,
        c='black'
    )

    ax.text(
        (x1 + x2)/2,
        y + h,
        text,
        ha='center',
        va='bottom',
        fontsize=14,
        fontweight='bold'
    )

# Position bracket above bars
y = max(means[0] + stds[0],
        means[1] + stds[1]) + 0.3

# Significance mark
add_significance(ax, x[0], x[1], y, 0.2, '***')

# Labels
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12)
ax.set_ylabel('GDT duration (h)', fontsize=16, labelpad=10)
ax.tick_params(axis='y', labelsize=12)

# Remove outer box
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Optional: make axes slightly thicker
ax.spines['left'].set_linewidth(1.2)
ax.spines['bottom'].set_linewidth(1.2)

# Set distance of bar from y-axis
ax.set_xlim(-0.5, 1.3)

# Set y-axis range
ax.set_ylim(0, 14)

plt.tight_layout()
plt.show()
```
